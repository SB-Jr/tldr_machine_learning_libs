# Understanding Data Returns

```python
import pandas as pd
from IPython.display import display

pd.options.display.max_columns = None

data_path = 'data/melb_data.csv'

data = pd.read_csv(data_path)

data.describe()
```

## Handling NA values

```python
cols_missing_value = [col for col in data.columns if data[col].isnull().any()]
print(cols_missing_value)
for col in cols_missing_value:
    print(str(col) + '-' + str(data[col].dtype))
```

### Drop columns <b>(Recomended)</b>

The easiest of all is to drop the column which has Nan values.

```python
reduced_data = data.drop(cols_missing_value, axis = 1)
reduced_data.head()
```

### Imputation (Recomended)

The issue with dropping the column is that we lose a lot of information which the column might provide to us. Here we try to fill in the missing values, one best way to do is by filling it with the mean. This might not look right, but it does yield a good result. 

```python
imputed_data = data.copy()
for col in cols_missing_value:
    if str(col) != 'CouncilArea':
        imputed_data[col] = imputed_data[col].fillna(data[col].mean())
        
imputed_data['CouncilArea'] = imputed_data['CouncilArea'].fillna('Unknown')

for col in ['Car']:#cols_missing_value:
    display(imputed_data[data[col].isnull()].head())
    display(data[data[col].isnull()].head())
```

As you can see many `NaN` values have been replaces in the `imputed_data` dataframe compared to the `data` dataframe.

### Extension to imputation

Another thing we can do is that we can add a new column which states whether the data has been added during preprocessing or was present in the dataset itself. We can add a Boolean column which states True for existing data and False for the rows, where we have added the mean value.

Note: We have to do this for each column for which we get a NaN valued row. 

```python
imputed_extension_data = data.copy()
for col in cols_missing_value:
    if col != 'CouncilArea':
        imputed_extension_data[col] = imputed_extension_data[col].fillna(data[col].mean())
        imputed_extension_data[col + '_was_missing'] = data[col].isnull()
imputed_extension_data['CouncilArea'] = imputed_extension_data['CouncilArea'].fillna('Unknown')
imputed_extension_data['CouncilArea_was_missing'] = data['CouncilArea'].isnull()


for col in ['Car']:#cols_missing_value:
    display(imputed_extension_data[data[col].isnull()].head())
    display(data[data[col].isnull()].head())
```

## Handling Categorical values

### Obtaining categorical columns

```python
labels = (data.dtypes == 'object')
categorical_columns = data.columns[labels]
display(categorical_columns)
## But for now we will only focus on Type, Method and Regionname
categorical_columns = ['Type', 'Method', 'Regionname']
```

### Category codes approach <b>(Recomended)</b>

```python
cat_data = data.copy()
cat_data[categorical_columns] = cat_data[categorical_columns].astype('category')
```

```python
for cat in categorical_columns:
    cat_data[cat + '_category'] = cat_data[cat].cat.codes
display(cat_data)
```

### Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
encoded_data = data.copy()
for cat in categorical_columns:
    encoded_data[cat + '_category'] = label_encoder.fit_transform(encoded_data[cat])
```

```python
display(encoded_data)
```

### One Hot Encoding

#### using encoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
one_hot_data = data.copy()
```

```python
oh_data = pd.DataFrame(one_hot_encoder.fit_transform(data[categorical_columns]))
oh_data.index = data.index
one_hot_data = one_hot_data.join(oh_data)
display(one_hot_data)
```

<b>Note: We need to join the new dataframe into our existing dataframe. Also make sure that the index is put into the new dataframe so that joining is easier</b>

#### Dummy apporach <b>(Recomended)</b>

```python
dummy_df = pd.get_dummies(data[categorical_columns], prefix=['Type_is', 'Method_is', 'Regionname_is'])
dummy_df.index = data.index
dummy_encoded_data = data.join(dummy_df)
```

```python
display(dummy_encoded_data)
```
